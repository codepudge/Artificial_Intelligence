```python
#divide into input and output
x= df.iloc[:,[0]].values
y= df.iloc[:,1].values
#x is taken in 2D always for sklearn and y is in 1D
```

```python
from sklearn.linear_model import LinearRegression as LR
#import sklearn.linear_model.LinearRegression as LR #another way of importing
```

```python
model= LR()
model.fit(x,y)  #creating the model
```

```python
y_pred = model.predict(x)
y_pred  #predicted op
```

```python
y #actual op
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[6.5]])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred)
```

```python
plt.figure(figsize=(8,4))
plt.scatter(x,y,c='k',marker="*",s=50,alpha=0.6,label="Actual value")
plt.plot(x,y_pred,c='r',label="Predicted value")
plt.xlabel("Years of Experience")
plt.ylabel("Salary Expected")
plt.title("Salary vs Experience")
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print(mean_squared_error(y,y_pred))
print(r2_score(y,y_pred))
```
