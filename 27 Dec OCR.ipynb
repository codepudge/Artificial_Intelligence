```python
import cv2
from google.colab.patches import cv2_imshow
import pytesseract as pt
img=cv2.imread('demo.jpg')
op=pt.image_to_string(img)
print(op)
cv2_imshow(img)
```

```python
h,w,d= img.shape
boxes=pt.image_to_boxes(img)
```

```python
print(boxes)
```

```python
for b in boxes.splitlines():
  b=b.split()
  x,y,w1,h1=int(b[1]),int(b[2]),int(b[3]),int(b[4])
  cv2.rectangle(img,(x,h-y),(w1,h-h1),(0,0,255),2)
  #cv2.putText(img,b[0],(x,h-y+25),cv2.FONT_HERSHEY_PLAIN,2,(0,0,0),1)
cv2_imshow(img)
```
