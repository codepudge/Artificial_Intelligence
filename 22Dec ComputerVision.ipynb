```python
img=cv2.imread('car.jpg')
cv2_imshow(img)
```

```python
img.shape
```

```python
#Resizing the image
img=cv2.resize(img,(int(1280/4),int(720/4)))
cv2_imshow(img)
```

```python
img.shape
```

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```python
kernel1= np.array([[1,0,-1],[0,0,0],[-1,0,1]])
op=cv2.filter2D(img_gray,-1,kernel1)
cv2_imshow(op)
```

```python
kernel2= np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
op=cv2.filter2D(img_gray,-1,kernel2)
cv2_imshow(op)
```

```python
#BGR to binary
b=cv2.threshold(img_gray,110,255,cv2.THRESH_BINARY)
cv2_imshow(b[1])
```

```python
import numpy as np
import cv2
from google.colab.patches import cv_imshow
```

```python
a= np.zeros([200,200,3],dtype="uint8")
cv2.line(a,(0,50),(100,50),(255,255,255),5)
cv2.line(a,(100,50),(100,150),(255,0,255),5)
cv2.line(a,(0,150),(100,150),(0,255,255),5)
cv2.line(a,(0,150),(0,50),(255,255,0),5)
cv2_imshow(a)
```

```python
a= np.zeros([200,200,3],dtype="uint8")
cv2.rectangle(a,(0,50),(100,100),(0,0,255),5)
cv2.putText(a,"Python",(50,150),cv2.FONT_HERSHEY_PLAIN,1,(0,255,0))
cv2_imshow(a)
cv2.imwrite("output.jpg",a)
```
